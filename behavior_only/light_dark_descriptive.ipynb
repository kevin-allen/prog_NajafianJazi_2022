```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from scipy import stats
withLetters=True
#sns.set_style("whitegrid")
plt.style.use("seaborn")
al=[1,0.6]

col=2
row=2
sizePerPlot=2.4
fig = plt.figure(figsize=(col*sizePerPlot,row*sizePerPlot))
gs = gridspec.GridSpec(row,col,
                       wspace=0.3, hspace=0.3) #
#fig.tight_layout()


#######################
### distributions #####
#######################

# distribution of trial duration
ax = plt.subplot(gs[0,0])
ax.hist(df.duration[df.light=="dark"],bins=np.arange(0,120,5),alpha=al[0],label="dark",color=sns.color_palette()[0])
ax.hist(df.duration[df.light=="light"],bins=np.arange(0,120,5),alpha=al[1],label="light",color=sns.color_palette()[4])
ax.set_xlabel("Duration (Sec)")
ax.set_ylabel("Trials")
ax.axvline(x=np.nanmedian(df.duration[df.light=="light"]),linestyle='dashed',color="black")
ax.axvline(x=np.nanmedian(df.duration[df.light=="dark"]),linestyle='dashed',color=sns.color_palette()[0])
#ax.set_xticks(np.arange(0,10,2))
ax.legend()
ax.yaxis.set_label_coords(-0.18, 0.5)
ax.xaxis.set_label_coords(0.5, -0.12)
ax.set_xticks(np.arange(0,150,50))
ax.tick_params(axis='both', which='major', pad=3)
stat,p_value = stats.mannwhitneyu(x=df.duration[df.light=="light"], y = df.duration[df.light=="dark"])
ax.text(0.45,0.6,"p = {:.02e}".format(p_value),transform=ax.transAxes)
ax.text(0.04,0.87, 'a', style='normal',fontsize=15, weight="bold", transform=plt.gcf().transFigure)
print("Duration mean:")
print("Light trials {}".format(np.nanmean(df.duration[df.light=="light"])))
print("Dark trials {}".format(np.nanmean(df.duration[df.light=="dark"])))
      


# number of journeys per trial
ax = plt.subplot(gs[0,1])
ax.hist(df.nJourneys[df.light=="dark"],bins=np.arange(0.5,10.5,1),alpha=al[0],label = "Dark",color=sns.color_palette()[0])
ax.hist(df.nJourneys[df.light=="light"],bins=np.arange(0.5,10.5,1),alpha=al[1],label="Light",color=sns.color_palette()[4])
ax.set_xlabel("Journeys")
ax.set_ylabel("Trials")
ax.axvline(x=np.nanmedian(df.nJourneys[df.light=="light"]),linestyle='dashed',color="black")
ax.axvline(x=np.nanmedian(df.nJourneys[df.light=="dark"]),linestyle='dashed',color=sns.color_palette()[0])
ax.set_xticks(np.arange(0,10,2))
ax.yaxis.set_label_coords(-0.18, 0.5)
ax.xaxis.set_label_coords(0.5, -0.12)
ax.tick_params(axis='both', which='major', pad=3)
stat,p_value = stats.mannwhitneyu(x= df.nJourneys[df.light=="light"], y = df.nJourneys[df.light=="dark"] )
ax.text(0.3,0.8,"p = {:.02e}".format(p_value),transform=ax.transAxes)
ax.text(0.475,0.87, 'b', style='normal',fontsize=15, weight="bold", transform=plt.gcf().transFigure)
print("Journeys per trial mean:")
print("Light trials {}".format(np.nanmean(df.nJourneys[df.light=="light"])))
print("Dark trials {}".format(np.nanmean(df.nJourneys[df.light=="dark"])))




# we will group per session to get an idea of the variability 
# probability that the lever is found on a given journey
dfJPress = df.loc[:,["light","nJourneys","nJourneysAtLever","nJourneysWithPress"]].groupby([df["light"],df["sessionName"]]).sum()
dfJPress=dfJPress.reset_index() # get rid of hierarchical index
dfJPress["findLeverProb"] = dfJPress["nJourneysAtLever"]/ dfJPress["nJourneys"]
dfJPress["pressProb"] = dfJPress["nJourneysWithPress"]/ dfJPress["nJourneysAtLever"]

# probability of finding the lever on a journey
ax = plt.subplot(gs[1,0])
ax.hist(dfJPress.findLeverProb[dfJPress.light=="dark"],label="Dark",bins=np.arange(0,1.05,0.05),alpha=al[0],color=sns.color_palette()[0])
ax.hist(dfJPress.findLeverProb[dfJPress.light=="light"],label="Light",bins=np.arange(0,1.05,0.05),alpha=al[1],color=sns.color_palette()[4])
ax.axvline(x=np.nanmedian(dfJPress.findLeverProb[dfJPress.light=="light"]),linestyle='dashed',color="black")
ax.axvline(x=np.nanmedian(dfJPress.findLeverProb[dfJPress.light=="dark"]),linestyle='dashed',color=sns.color_palette()[0])
ax.set_xlim(0,1)
ax.set_xlabel("Finding lever (prob.)")
ax.set_ylabel("Sessions")
ax.tick_params(axis='both', which='major', pad=4)
stat,p_value = stats.mannwhitneyu(x= dfJPress.findLeverProb[dfJPress.light=="light"], y = dfJPress.findLeverProb[dfJPress.light=="dark"])
ax.text(0.05,0.8,"p = {:.02e}".format(p_value),transform=ax.transAxes)
ax.text(0.04,0.44, 'c', style='normal',fontsize=15, weight="bold", transform=plt.gcf().transFigure)
print("Probability of finding the lever mean")
print("Light trials {}".format(np.nanmean(dfJPress.findLeverProb[dfJPress.light=="light"])))
print("Dark trials {}".format(np.nanmean(dfJPress.findLeverProb[dfJPress.light=="dark"])))





# probability of pressing the lever when the lever is found
ax = plt.subplot(gs[1,1])

ax.hist(dfJPress.pressProb[dfJPress.light=="dark"],label="Dark",bins=np.arange(0,1.05,0.05),alpha=al[0],color=sns.color_palette()[0])
ax.hist(dfJPress.pressProb[dfJPress.light=="light"],label="Light",bins=np.arange(0,1.05,0.05),alpha=al[1],color=sns.color_palette()[4])
ax.axvline(x=np.nanmedian(dfJPress.pressProb[dfJPress.light=="light"]),linestyle='dashed',color="black")
ax.axvline(x=np.nanmedian(dfJPress.pressProb[dfJPress.light=="dark"]),linestyle='dashed',color=sns.color_palette()[0])
ax.set_xlim(0,1)
ax.set_xlabel("Pressing lever (prob.)")
ax.set_ylabel("Sessions")
ax.tick_params(axis='both', which='major', pad=4)
stat,p_value = stats.mannwhitneyu(x=dfJPress.pressProb[dfJPress.light=="light"] , y =dfJPress.pressProb[dfJPress.light=="dark"] )
ax.text(0.05,0.8,"p = {:.02e}".format(p_value),transform=ax.transAxes)
ax.text(0.475,0.44, 'd', style='normal',fontsize=15, weight="bold", transform=plt.gcf().transFigure)
print("Probability of pressing the lever at the lever mean")
print("Light trials {}".format(np.nanmean(dfJPress.pressProb[dfJPress.light=="light"])))
print("Dark trials {}".format(np.nanmean(dfJPress.pressProb[dfJPress.light=="dark"])))


fileName = myProject.dataPath+"/results/light_dark_descriptive_02.png"
print("Saving to " + fileName)
print("")
plt.savefig(fileName,bbox_inches = "tight",dpi=300)
```

## Prediction of light condition using 8 trial features using SVM model

I thought it would be interesting to show that the paths during light and dark trials are really different and that we can classify the path as light and dark with a very high degree of accuracy.

```python
# standardize the feature so that they have a similar weight in our model
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_score

# extract the data from the df data frame
vars = ["length","duration","meanSpeed","complexity"]
paths = ["searchArenaNoLever" ,"homingPeriNoLever"]
vp = [p+"_"+v for v in vars for p in paths] # variables that are the predictors
#allV = vp + ["lightInt"]

# shuffle df
df = df.sample(frac = 1)


y = (df["light"]=="light").astype(int)
df["searchArenaNoLever_meanSpeed"] = np.where(df["searchArenaNoLever_meanSpeed"]>100,np.nan,df["searchArenaNoLever_meanSpeed"])



## select the variables of interest
class dataFrameSelector(BaseEstimator, TransformerMixin):
    """
    Take a DataFrame in and output a np.array
    """
    def __init__(self, variables):
        self.variables = variables
    def fit(self,X,y=None):
        return self # nothing to fit
    def transform(self, X, y = None):
        return X[self.variables].values


class varLogDrop(BaseEstimator, TransformerMixin):
    """
    transform some variable by applying log transformation
    drop rows with NAs
    """
    def __init__(self, logIndices):
        self.logIndices = logIndices
    def fit(self,X,y=None):
        return self # nothing to fit
    def transform(self, X, y = None):
        # log
        for i in self.logIndices:
            X[:,i] = np.log1p(X[:,i])
        # drop nan
        #X = X[~np.isnan(X).any(axis=1)]
        return X

myPipeline = Pipeline([
    ('dfs',dataFrameSelector(vp)),
    ('vld', varLogDrop([0,1,2,3])),
    ('imputer',SimpleImputer()),
    ('std_scaler',StandardScaler())])

XPrep = myPipeline.fit_transform(df)
print("input shape")
print(XPrep.shape,y.shape)

sgd_clf = SGDClassifier()
res = cross_val_score(sgd_clf, XPrep,y,cv=12,scoring="accuracy")
print(res)
print("Mean accuracy of our model: {}, std: {}".format(np.mean(res), np.std(res)))
```
