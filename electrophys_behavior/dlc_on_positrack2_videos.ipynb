```python
from positrack2.offline_tracker_dlc import OfflineTrackerDlc
```

```python
def positrack2_offline(sSes,model_config_file = "/adata/models/positrack2_150_150_2LEDs-Allen-2021-07-14/config.yaml"):
    """
    Function to run dlc on the video saved by positrack2
    
    Assumes that there is a trialName.mp4 file generated by positrack2
    Takes approximately 10 minutes per trial (100 Hz)
    
    Arguments
    
    sSes: spikeA session
    config_file: config file of the dlc model you want to use
    
    """
    input_video_file_list = [ "{}/{}.mp4".format(sSes.path,fn) for fn in sSes.trial_names]
    output_tracking_file_list = [ "{}/{}.offline.positrack2".format(sSes.path,fn) for fn in sSes.trial_names]
    output_video_file_list = [ "{}/{}_labeled.avi".format(sSes.path,fn) for fn in sSes.trial_names]
    
    for vidi, trk, vido in zip(input_video_file_list,output_tracking_file_list, output_video_file_list):
        print(vidi,trk,vido)
    
        OfflineTrackerDlc(model_config_file = model_config_file,
                          input_video_file= vidi,
                          output_tracking_file= trk,
                          output_video_file= vido)
```
