Explain how we rotated the data around the lever.

We assumes that the lever is at position 0,0. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Mouse position around the lever

```python
mouse = np.array([30,0])
mouse
```

Bridge position around the lever

```python
bridge = np.array([4,-40])
```

Let's assume the lever is at the center of the arena. Just to plot a circle that can help see the rotation we are applying.

```python
arenaX = np.cos(np.linspace(0,np.pi*2,100))*40
arenaY = np.sin(np.linspace(0,np.pi*2,100))*40
```

```python
plt.plot(arenaX,arenaY)
plt.scatter(mouse[0],mouse[1])
plt.scatter(bridge[0],bridge[1],color="red")
```

You can get the angle of any 2D vector with np.arctan2. Watch out for the y,x order. This is the angle relative to the east (1,0)

```python
bridgeAngle = np.arctan2(bridge[1],bridge[0])
bridgeAngle, -np.pi/2
```

We add np.pi/2 to the angle. This gives us the bridge angle relative to a vector poining south.

```python
rotation = bridgeAngle+np.pi/2
rotation
```

```python
rotation*180/np.pi
```

To rotate our position data, we use a rotation matrix. 

```python
rotMat = np.array([[np.cos(rotation), -np.sin(rotation)],
                       [np.sin(rotation), np.cos(rotation)]])
rotMat
```

Apply the rotation to our position data to rotate them around 0,0

```python
newMouse = mouse@rotMat
newBridge = bridge@rotMat
```

```python
plt.plot(arenaX,arenaY)
plt.scatter(mouse[0],mouse[1])
plt.scatter(bridge[0],bridge[1],color="red")
plt.scatter(newMouse[0],newMouse[1],color="blue", alpha=0.5)
plt.scatter(newBridge[0],newBridge[1],color="red",alpha=0.5)
```
